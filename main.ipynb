```python
import pandas as pd
from dotenv import load_dotenv
import os
```

```python
# Set environment variables from the .env in the local environment
load_dotenv()

url_2015 = os.getenv("url_2015")
url_2014 = os.getenv("url_2014")
url_2013 = os.getenv("url_2013")
url_2012 = os.getenv("url_2012")
url_2011 = os.getenv("url_2011")
```

```python
# 2015
url_2015 = url_2015.replace('www.dropbox.com', 'dl.dropboxusercontent.com')
df_2015 = pd.read_csv(url_2015)
df_2015.head()
```

```python
# 2014
url_2014 = url_2014.replace('www.dropbox.com', 'dl.dropboxusercontent.com')
df_2014 = pd.read_csv(url_2014)   
df_2014.head()
```

```python
# 2013
url_2013 = url_2013.replace('www.dropbox.com', 'dl.dropboxusercontent.com')
df_2013 = pd.read_csv(url_2013)   
df_2013.head()
```

```python
# 2012
url_2012 = url_2012.replace('www.dropbox.com', 'dl.dropboxusercontent.com')
df_2012 = pd.read_csv(url_2012)   
df_2012.head()
```

```python
# 2011
url_2011 = url_2011.replace('www.dropbox.com', 'dl.dropboxusercontent.com')
df_2011 = pd.read_csv(url_2011)   
df_2011.head()
```

```python
# Create sets of the columns
set1 = set(df_2015.columns)
set2 = set(df_2014.columns)
set3 = set(df_2013.columns)
set4 = set(df_2012.columns)
set5 = set(df_2011.columns)

# Find the intersection of the sets to get common columns
common_columns = list(set1.intersection(set2).intersection(set3).intersection(set4).intersection(set5))

# Combine the dataframes using the common columns
combined_df = pd.concat([df_2015[common_columns], df_2014[common_columns], df_2013[common_columns], df_2012[common_columns], df_2011[common_columns]])
combined_df.head()
```

```python
combined_df.shape
```

```python
combined_df.sort_values(by="IDATE", ascending=False, inplace=True)
combined_df.reset_index(drop=True, inplace=True)
combined_df.shape
```

```python
# Function to clean string
def clean_string(string_value):
    cleaned_value = string_value.replace("b'", "").replace("'", "").strip() 
    return cleaned_value

# Apply the clean function to the 'IYEAR' column
combined_df["IYEAR"] = combined_df['IYEAR'].apply(clean_string)
combined_df["IYEAR"].head()
```

```python
# Apply the clean function to the 'IMONTH' column
combined_df["IMONTH"] = combined_df['IMONTH'].apply(clean_string)
combined_df["IMONTH"].head()
```

```python
combined_df['month_year'] = pd.to_datetime(combined_df['IYEAR'].astype(str) + '-' + combined_df['IMONTH'].astype(str), format='%Y-%m')

```

```python
combined_df.drop(columns=['IYEAR', 'IMONTH', 'IDAY', "IDATE"], inplace=True)
```
